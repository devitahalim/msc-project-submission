```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
dataDirectory = '/Users/devitahalim/Documents/GitHub/VRPT_Project/data/'
instanceList = ['AuPost_Northern_2019','AuPost_Central_2019','AuPost_Western_2019', 'AuPost_Southern_2019'] 

instanceList = ['AuPost_Northern_2019_Mixed','AuPost_Central_2019_Mixed','AuPost_Western_2019_Mixed', 'AuPost_Southern_2019_Mixed'] 

for file_name in instanceList[1:2]:
    case = file_name.split('_')[1]
    df = pd.read_excel(dataDirectory+ file_name + '.xlsx', sheet_name='Demand(items)')
    df = df.iloc[6:9]
    plt.figure(figsize=(7, 6))  # Adjust the figure size as needed
    plt.rcParams.update({'font.size': 14})
    store_names = pd.read_excel(dataDirectory+ file_name + '.xlsx', sheet_name=case+' SA4 nodes', usecols = ['SA4'])
    df.index = store_names['SA4'].tolist()
    df.columns = df.index

    store_names = ['1', '2', '3']
    df.index = store_names
    df.columns = df.index
    sns.heatmap(df, cmap = 'Purples', annot=False, fmt='.1f', linewidths=.5, cbar_kws={'shrink': 0.8}, square=True)
    # plt.xticks(fontsize=8)
    # plt.yticks(fontsize=8)
    plt.xlabel('Origin')
    plt.ylabel('Destination')
    # plt.title(f'{case}')
    # plt.subplots_adjust(left=0.1, right=1.0, top=0.9, bottom=0.3)

    # Show the plot
    plt.show()
    case = file_name.split('_')[1]
    df = pd.read_excel(dataDirectory+ file_name + '.xlsx', sheet_name='Demand(items)')
    df = df.iloc[12:15]
    plt.figure(figsize=(7, 6))  # Adjust the figure size as needed
    plt.rcParams.update({'font.size': 14})
    store_names = pd.read_excel(dataDirectory+ file_name + '.xlsx', sheet_name=case+' SA4 nodes', usecols = ['SA4'])
    df.index = store_names['SA4'].tolist()
    df.columns = df.index

    store_names = ['1', '2', '3']
    df.index = store_names
    df.columns = df.index
    sns.heatmap(df, cmap = 'Purples', annot=False, fmt='.1f', linewidths=.5, cbar_kws={'shrink': 0.8}, square=True)
    # plt.xticks(fontsize=8)
    # plt.yticks(fontsize=8)
    plt.xlabel('Origin')
    plt.ylabel('Destination')
    # plt.title(f'{case}')
    # plt.subplots_adjust(left=0.1, right=1.0, top=0.9, bottom=0.3)

    # Show the plot
    plt.show()
    
```
